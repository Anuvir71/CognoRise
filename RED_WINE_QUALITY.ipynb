Exploratory Data Analysis


```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['quality'],y=df['volatile acidity'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=df['pH'],y=df['fixed acidity'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(x=df['citric acid'],bins=15,edgecolor='black')
plt.xlabel("citric acid")
plt.ylabel("level")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="residual sugar",y="alcohol",data=df)
plt.ylabel("alcohol",fontsize=11)
plt.title("level of residual sugar",fontsize=11)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,5))
ax=sns.countplot(x=df['alcohol'])
for bars in ax.containers:
  ax.bar_label(bars)
  plt.xticks(rotation=90)
  plt.show()
```

```python
sns.scatterplot(x=df['alcohol'],y=df['quality'])
plt.show()
```

```python
sns.regplot(x=df['free sulfur dioxide'],y=df['quality'])
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x=df['citric acid'],y=df['density'])
plt.xticks(rotation=90)
plt.show()
```
